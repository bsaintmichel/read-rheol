## Main Script

This script reads a .csv file from Trios (TA Instrument rheometers, so AR-G2, ARES, DHR, ...) and spits out three objects :
- `data` : a `pandas.DataFrame` containing all variables for each individual step
- `units` : a `dict` containing the units of all the corresponding columns of `data`, plus some of that contained in the metadata (e.g. instrument inertia, ...)
- `info` : another `dict` containing experiment metadata (gap size, instrument inertia, ...). For now, not necessarily in useable units, but can easily be fixed.

NOTE : for now, this version of the code works only with Trios, the aim is to (later on) have it work in the same way for Anton Paar and Malvern rheometers.

```python
import rheol_functions as rh
import matplotlib.pyplot as plt

# Load either file
file = './sample_Trios_2.csv'
# file = './Carbopol_AntonPaar.csv'
# file = './Carbopol_Malvern.csv'
# file = 'Z:/Kinexus/Carbopol_20220625_0.1pc_Omar.csv'

data, units, info = rh.read_trios(file, delim=',')
data, units = rh.format_trios(data, units, info)
```

#### Plotting Flow Sweep Data

Also known as flow curves by anybody else working on the subject... Anyway. The following code is very simple.

```python
flow_curve = data[data['name'] == 'Flow sweep']

fig, ax = plt.subplots()
ax.loglog(flow_curve['shear_rate'], flow_curve['stress'], '+')
ax.set_ylim([0.1,100])
ax.set_xlabel('Shear Rate (1/s)')
ax.set_ylabel('Stress (Pa)')
```

#### Plotting amplitude sweep data

Very similar syntax compared to plotting the flow curves.

```python
flow_curve = data[data['name'] == 'Amplitude sweep']

fig, ax = plt.subplots()

ax.loglog(flow_curve['osc_strain'], flow_curve['g_prime'], '+', label='G\'')
ax.loglog(flow_curve['osc_strain'], flow_curve['g_second'], 'rx', label='G\'\'')

ax.set_xlabel('Strain (%)')
ax.set_ylabel('Shear Modulus (Pa)')
ax.set_ylim([0.1,100])
ax.legend()
```
